Basic SVM with linear kernal 

EX 4.2



```python
import sys,os
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score


x,y = make_blobs(n_samples=10000,centers=2,n_features=4)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=137)

model = svm.SVC(kernel='linear',C=200).fit(X_train,y_train)
y_pred = model.predict(X_test)

acc = accuracy_score(y_test,y_pred)
pre = precision_score(y_test,y_pred,average='micro')
rec = recall_score(y_test,y_pred,average='micro')
print("accuracy precision recall")
print(acc,     pre,     rec)
```

SVM on breast cancer dataset

```python
import sys,os
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score


d = load_breast_cancer()
x = d.data
y = d.target
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=137)

model = svm.SVC(kernel='linear',C=200).fit(X_train,y_train)
y_pred = model.predict(X_test)

acc = accuracy_score(y_test,y_pred)
pre = precision_score(y_test,y_pred)
rec = recall_score(y_test,y_pred)
print("accuracy precision recall")
print(acc,     pre,     rec)
```
